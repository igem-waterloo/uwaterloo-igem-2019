```julia
using DiffEqBiological;
using DifferentialEquations;
using Plots;
using Latexify;
using NumericIO;
```

```julia
const Avogadro = 6.022140857e23;
const kda_mg_conversion = 1.66054e-18;
```

```julia
const PuhB_K_m = 7600; # nmol
const PuhB_k_cat = 258; # 1 / s, error is plus or minus 8
const NAT1_k = 68; # nmol/(min*mg), error is plus or minus 8
const NAT1_weight = 33; # kDa
```

```julia
rn_influx = @reaction_network begin
    l, 0 --> X
    mm(X, v, K), X --> Y
    k, Y --> Z
    end l v K k;
```

```julia
latexify(rn_influx)
```

```julia
NAT1_conc = 1*10^(-12); # mol
NAT1_conc_factor = (NAT1_conc*Avogadro)*(NAT1_weight*kda_mg_conversion); # mg
NAT1_rate = NAT1_k*NAT1_conc_factor; # 1/min

PuhB_conc = 1*10^(-14); # mol
PuhB_V_max = (PuhB_conc*Avogadro*PuhB_k_cat)*(10.0)^(-(23-9)) ; # nmol/min

Linuron_influx_rate = 0.5; # mol

p = (Linuron_influx_rate, PuhB_V_max, PuhB_K_m, NAT1_rate);
u0 = [600, 0.001, 0.001]; # initial concentrations in nmol
tspan = (0., 1000.);

prob = ODEProblem(rn_influx, u0, tspan, p);
sol = solve(prob, DynamicSS(Tsit5()));
plot(sol)
```

```julia
u0 = [200., 200., 1800.]
ssprob = SteadyStateProblem(rn_influx, u0, p)
sssol  = solve(ssprob, SSRootfind())[1:2]
```
